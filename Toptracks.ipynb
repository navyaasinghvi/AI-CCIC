# What genre appeared most frequently in Spotify's "Top tracks of 2024" USA?


This will be an observational study because no variables are being changed-- you are just collecting samples and recording historical data.
I will be studying the entire population, meaning the entire playlist, to make sure my data is accurate enough to answer what language is most popular. The study is also known as a census. The dataset is based on Spotify’s playlist for the USA titled "Top Tracks of 2024 (USA)." Spotify, the music streaming platform, is the primary data collector. This data set however, is borrowed from my teacher, who also used the help of AI to make it. Spotify uses internal analytics to track total global and regional streams. The algorithm usually includes the play count (streams), the number of likes, shares, or even saves, how much time the user spent listening to the song, and how popular it is. Spotify only included tracks that made the top stream counts for the region. Lesser-streamed songs, remixes, niche genre tracks, or independent releases may not appear. One of the many biases may include selection bias—the dataset includes only the most-streamed tracks on Spotify USA in 2024, which excludes all songs with lower streams or those not on Spotify or not popular in the USA. The sample is not representative of all songs released in 2024, only the "popular" subset. Sure! This makes some songs or artists seem more popular than they are.**




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('toptracks.csv')

# Create frequency table of the 'AI Genre' column
genre_counts = df['AI Genre'].value_counts()

print("Frequency Table of Genres:")
print(genre_counts)

# Visualize the frequency table with a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')

plt.title("Frequency of Genres in Spotify's Top Tracks of 2024 (USA)")
plt.xlabel("Genre")
plt.ylabel("Number of Tracks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('toptracks.csv')

# Create relative frequency table
genre_rel_freq = df['AI Genre'].value_counts(normalize=True)

# Print the relative frequency table
print("Relative Frequency Table of 'AI Genre':")
print(genre_rel_freq)

# Visualize the relative frequency
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_rel_freq.index, y=genre_rel_freq.values, palette='viridis')

plt.title("Relative Frequency of Genres in Spotify's Top Tracks of 2024 (USA)")
plt.xlabel("Genre")
plt.ylabel("Proportion of Tracks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


One conclusion you can make from this data is that Pop is the most dominant genre in Spotify’s Top Tracks of 2024 (USA), making up 32% of the songs, followed by Country at 22% and Hip-Hop at 18%. This shows that these three genres alone account for over 70% of the top tracks, while other genres like Indie Rock, Pop Rock, and Alternative Pop each make up only 2%, showing they are much less common. This suggests that mainstream listeners are heavily favoring Pop, Country, and Hip-Hop music in 2024.
