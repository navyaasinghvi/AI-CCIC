How has the average price of Amazon’s bestselling books changed from 2009 to 2019, and what relationship exists between book price and user rating?

# My study- 
This is an observational study that looks at trends in Amazon’s bestselling books from 2009 to 2019. The data was colleThis is an observational study that looks at trends in Amazon’s bestselling books from 2009 to 2019. The dataset includes books that appeared on Amazon’s yearly bestseller lists, not every book sold on the site. It was collected from Amazon’s public listings and contains information such as the book’s title, author, user rating, number of reviews, price, year, and genre. Since it only includes top-selling books, there may be some bias because it focuses on popular and highly advertised titles rather than smaller or lesser-known ones.

In this study, descriptive statistics, correlation analysis, and graphs are used to find out how book prices have changed over time and whether there is any connection between a book’s price and its user rating. The goal is to see if higher prices are linked to better ratings and to understand general patterns in book pricing and reader satisfaction over the decade. While the results do not represent all books on Amazon, they help show how bestselling books have performed and been received over the years.analysis, and graphs to see if book prices changed over time and to find out if there is any connection between a book’s price and its user rating.

```python
# Amazon Bestseller Analysis (2009–2019)
# Research Question:
# How has the average price of Amazon’s bestselling books changed from 2009 to 2019,
# and what relationship exists between price and user rating?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("books.csv")

# Basic exploration
df.head()
df.info()
df.isnull().sum()
len(df)

# Descriptive statistics
df[['User Rating', 'Reviews', 'Price']].describe()

# Central tendency values
mean_price = df['Price'].mean()
median_price = df['Price'].median()
mean_rating = df['User Rating'].mean()

# --- Price Distribution with IQR, Mean, and Median ---
plt.figure(figsize=(10, 6))

# Distribution plot for Price
sns.histplot(df['Price'], bins=30, kde=True, color='skyblue', edgecolor='black')

# Compute quartiles and IQR
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3 - q1
mean_price = df['Price'].mean()
median_price = df['Price'].median()

# Add vertical lines for quartiles, mean, and median
plt.axvline(q1, color='orange', linestyle='--', label=f'Q1 (25%): ${q1:.2f}')
plt.axvline(q3, color='orange', linestyle='--', label=f'Q3 (75%): ${q3:.2f}')
plt.axvline(mean_price, color='red', linestyle='-', label=f'Mean: ${mean_price:.2f}')
plt.axvline(median_price, color='green', linestyle='-', label=f'Median: ${median_price:.2f}')

# Add IQR annotation
plt.text(q3 + 0.5, plt.ylim()[1]*0.9, f'IQR = ${iqr:.2f}', color='purple', fontsize=10)

plt.title("Distribution of Amazon Bestseller Prices (2009–2019)")
plt.xlabel("Book Price ($)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# --- Average price trend over time ---
avg_price_by_year = df.groupby('Year')['Price'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=avg_price_by_year, x='Year', y='Price', marker='o')
plt.title("Average Price of Amazon Bestsellers (2009–2019)")
plt.xlabel("Year")
plt.ylabel("Average Price ($)")
plt.grid(True)
plt.show()

# --- Correlation between Price and User Rating ---
correlation = df['Price'].corr(df['User Rating'])
print(f"Correlation between Price and User Rating: {correlation:.3f}")

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Price', y='User Rating', data=df, alpha=0.6)
plt.title("Relationship Between Book Price and User Rating")
plt.xlabel("Book Price ($)")
plt.ylabel("User Rating (0–5 Stars)")
plt.grid(True)
plt.show()

# --- Genre distribution ---
plt.figure(figsize=(10, 6))
sns.countplot(y='Genre', data=df, order=df['Genre'].value_counts().index)
plt.title("Number of Bestsellers by Genre (2009–2019)")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

# My answer- 
Based on the data from 2009 to 2019, the average price of Amazon’s bestselling books showed only small year-to-year variations, remaining relatively stable overall. The mean price was about $13, while the median price was around $11, suggesting that most bestsellers were moderately priced and that a few higher-priced books slightly increased the average.

The first quartile (Q1) was roughly $8, and the third quartile (Q3) was around $16, giving an interquartile range (IQR) of about $8. This means that half of all bestseller prices fell between $8 and $16.

The average user rating across all books was approximately 4.6 out of 5, showing that bestselling books consistently received high ratings from readers. The correlation between price and user rating was very close to zero, indicating that a book’s price had little to no impact on how highly it was rated.

Overall, the data shows that between 2009 and 2019, Amazon’s bestselling books maintained steady prices and consistently high ratings, with no strong relationship between cost and customer satisfaction.
