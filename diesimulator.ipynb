```python
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Simulate 100 die rolls and print each one
def roll_die(sides=6, trials=100):
    results = []

    print(f"Rolling a {sides}-sided die {trials} times:\n")

    for trial in range(1, trials + 1):
        roll = random.randint(1, sides)
        print(f"Trial {trial}: Rolled a {roll}")
        results.append(roll)

    df = pd.DataFrame(results, columns=['Roll'])
    return df

# Step 2: Roll the die and print trials
df_rolls = roll_die()

# Step 3: Count frequencies
counts = df_rolls['Roll'].value_counts().sort_index()
df_counts = counts.reset_index()
df_counts.columns = ['Die Face', 'Frequency']

# Step 4: Visualize with seaborn
sns.set(style='whitegrid')
sns.barplot(data=df_counts, x='Die Face', y='Frequency', palette='muted')

plt.title('Results of Rolling a 6-Sided Die 100 Times')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.show()
```
