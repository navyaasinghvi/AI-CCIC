# Question: 
The average IMDb rating of movies in a dataset is 7.27, with a standard deviation of 1.03.
What is the probability that a randomly selected movie has a rating between 7.5 and 9.0?

# Study: 

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# --- Example IMDb ratings dataset ---
ratings = [6.2, 7.8, 5.5, 8.1, 7.3, 6.8, 9.0, 7.9, 8.4, 5.9, 6.5, 7.2, 8.6, 6.7, 7.1]

# --- 1️⃣ Calculate mean and standard deviation ---
mean = np.mean(ratings)
std_dev = np.std(ratings, ddof=1)  # sample standard deviation

print(f"Mean IMDb Rating: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

# --- 2️⃣ Define interval ---
lower = 7.5
upper = 9.0

# --- 3️⃣ Theoretical Probability (using normal distribution) ---
theoretical_prob = norm.cdf(upper, mean, std_dev) - norm.cdf(lower, mean, std_dev)
print(f"Theoretical Probability (7.5 ≤ X ≤ 9.0): {theoretical_prob:.4f}")

# --- 4️⃣ Empirical Probability (based on actual data) ---
empirical_prob = np.mean((np.array(ratings) >= lower) & (np.array(ratings) <= upper))
print(f"Empirical Probability (7.5 ≤ X ≤ 9.0): {empirical_prob:.4f}")

# --- 5️⃣ Visualization ---
x = np.linspace(min(ratings) - 1, max(ratings) + 1, 1000)
pdf = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))

# Histogram (empirical distribution)
plt.hist(ratings, bins=8, density=True, alpha=0.6, color='skyblue', label='Empirical Ratings')

# Theoretical normal curve
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical Normal PDF')

# Highlight interval 7.5–9.0
x_fill = np.linspace(lower, upper, 200)
plt.fill_between(x_fill, norm.pdf(x_fill, mean, std_dev), color='orange', alpha=0.4,
                 label=f'Interval [{lower}, {upper}]')

plt.title("IMDb Ratings: Theoretical vs Empirical Distribution")
plt.xlabel("Rating")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

# Answer: 
Based on the code results, the average IMDb rating of the dataset was 7.27 with a standard deviation of 1.03. The theoretical probability that a randomly selected movie has a rating between 7.5 and 9.0 was 0.2774 (27.74%), assuming the data follows a normal distribution. The empirical probability calculated directly from the dataset was 0.4000 (40%), meaning that 6 out of 15 movies actually fall within that range. This shows that while the normal model predicts around 28% of movies to have ratings between 7.5 and 9.0, the real data suggests a higher concentration of highly rated movies.
